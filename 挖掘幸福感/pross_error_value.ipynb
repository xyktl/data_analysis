```python
import numpy as np
import pandas as pd
import tensorflow as tf
import random
```

```python
def import_data(path):
    try:
        data = pd.read_csv(path)
        return data
    except:
        raise IOError
```

```python
def unusl_value_process(data):
    #处理空值
    data["work_status"].fillna(0, inplace=True)
    data["work_yr"].fillna(0,inplace=True)
    data["work_type"].fillna(0,inplace=True)
    data["work_manage"].fillna(0,inplace=True)
    data["family_income"].fillna(data["family_income"].mean(),inplace=True)
    
    #处理负数
    #happiness
    data.drop(list(data[data["happiness"]<0].index),inplace=True)
    #nationlity   按概率产生整数
    p = np.array([0.6,0.05,0.05,0.05,0.05,0.05,0.05,0.1])
    i = np.random.choice([1,2,3,4,5,6,7,8], p = p.ravel())
    data.loc[data[data["nationality"]<0].index,"nationality"] = i  
    
    #religion
    a1 = data[data["religion"]==0]["religion"].value_counts().sum() / data[data["religion"]>=0]["religion"].value_counts().sum()
    a = np.array([a1,1-a1])
    data.loc[data[data["religion"]<0].index,"religion"] = np.random.choice([0,1],p=a.ravel())
    
    #religion_freq
    data.drop(list(data[data["religion_freq"]<0].index),inplace=True)
    #edu:
    data.drop(list(data[data["edu"]<0].index),inplace=True)
    #income
    data.loc[data[data["income"]<0].index,"income"] = data["income"].mean()
    #political
    p1 = data[data["political"]==1]["political"].value_counts().sum() / data[data["political"]>0]["political"].value_counts().sum()
    p2 = data[data["political"]==2]["political"].value_counts().sum() / data[data["political"]>0]["political"].value_counts().sum()
    p3 = data[data["political"]==3]["political"].value_counts().sum() / data[data["political"]>0]["political"].value_counts().sum()
    #p4 = data[data["political"]=4]["political"].value_counts().sum() / data[data["political"]>=0]["political"].value_counts.sum()
    p = np.array([p1,p2,p3,(1-p1-p2-p3)])
    data.loc[data[data["political"]<0].index,"political"] = np.random.choice([1,2,3,4],p=p.ravel())
    #health
    data.loc[data[data["health"]<0].index,"health"] = random.randint(1,5)
    #health_problem
    data.loc[data[data["health_problem"]<0].index,"health_problem"] = random.randint(1,5)
    #depression
    data.loc[data[data["depression"]<0].index,"depression"] = random.randint(1,5)
    #socialize
    data.loc[data[data["socialize"]<0].index,"socialize"] = random.randint(1,5)
    #relax
    data.loc[data[data["relax"]<0].index,"relax"] = random.randint(1,5)
    #learn
    data.loc[data[data["learn"]<0].index,"learn"] = random.randint(1,5)
    #equity
    data.loc[data[data["equity"]<0].index,"equity"] = random.randint(1,5)
    #class
    data.loc[data[data["class"]<0].index,"class"] = random.randint(1,10)
    #family_m
    data.loc[data[data["family_m"]<0].index,"family_m"] = random.randint(1,4)
    #family_status
    data.loc[data[data["family_status"]<0].index,"family_status"] = random.randint(1,5)
    #house
    h = np.array([0.6,0.2,0.1,0.05,0.05])
    data.loc[data[data["house"]<0].index,"house"] = np.random.choice([1,2,3,4,5],p=h.ravel())
    #car
    c = np.array([0.2,0.3,0.5])
    data.loc[data[data["car"]<0].index,"car"] = np.random.choice([0,1,2],p=c.ravel())
    #status_peer
    data.loc[data[data["status_peer"]<0].index,"status_peer"] = random.randint(1,3)
    #status_3_before
    data.loc[data[data["status_3_before"]<0].index,"status_3_before"] = random.randint(1,3)
    #view
    data.loc[data[data["view"]<0].index,"view"] = random.randint(1,5)
    #inc_ability
    i1 = data[data["inc_ability"]==1]["inc_ability"].value_counts().sum() / data[data["inc_ability"]>=1]["inc_ability"].value_counts().sum()
    i2 = data[data["inc_ability"]==2]["inc_ability"].value_counts().sum() / data[data["inc_ability"]>=1]["inc_ability"].value_counts().sum()
    i3 = data[data["inc_ability"]==3]["inc_ability"].value_counts().sum() / data[data["inc_ability"]>=1]["inc_ability"].value_counts().sum()
    #i4 = data[data["inc_ability"]=4].value_counts().sum() / data["inc_ability"].value_counts().sum()
    i = np.array([i1,i2,i3,(1-i1-i2-i3)])
    data.loc[data[data["inc_ability"]<0].index,"inc_ability"] = np.random.choice([1,2,3,4],p=i.ravel())
```
